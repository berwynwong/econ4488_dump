```python
import scipy.stats as scs


def categories(series):
    return range(int(series.min()), int(series.max()) + 1)


def chi_square_of_df_cols(df, col1, col2):
    df_col1, df_col2 = df[col1].cat.as_ordered(), df[col2].cat.as_ordered()

    result = [[sum((df_col1 == cat1) & (df_col2 == cat2))
               for cat2 in categories(df_col2)]
              for cat1 in categories(df_col1)]

    return scs.chi2_contingency(result)
```

```python
def median(dfvar):
    x = dfvar.unique()
    cat_no = len(x)
    count = 0
    for i in range(cat_no):
        count += dfvar.value_counts(normalize=True)[i]
        if count >0.5 :
            return i
            break
        else:
            pass
```

```python
varlist1=['race', 'edu', 'age', 'indgen', 'own', 'marst', 'urb', 'claswk', 'geo2', 'birstate']
def central(df, varlist=varlist1):
    ave_list = np.empty(len(varlist))
    for i, var in enumerate(varlist):
        if var == 'edu':
            ave = median(df.edu)
        elif df[var].dtype.name =='category':
            ave = df[var].mode()[0]
        else:
            ave = df[var].median()
        ave_list[i] = ave
    return dict(zip(varlist, ave_list))
```

```python
def average_characteristic(df, j, varlist):
    df_split = df_race(df, j)
    x = models(df_split, reg_full)
    idx = x.params.index
    char = [1]
    for var in varlist:
        if df[var].dtype.name =='category':
            var_array = np.array(df_split[var].value_counts(sort=False, normalize=True))
            char = np.concatenate((char, var_array[1:]), axis=0)
        elif var == 'age':
            var_mean = df_split['age'].mean()
            char = np.concatenate((char, [var_mean, np.power(var_mean, 2)]), axis=0)
    return pd.Series(char, index=idx)
```

```python
def oaxaca_blinder_simple(df, races=[0,1], varlist=regressors, reg_form=reg_full):
    x_list = []
    d_list = []
    beta_list = []
    for j in races:
        ave_char = np.array(average_characteristic(df, j, varlist))
        df_split = df_race(df, j)
        reg_model = models(df_split, reg_form)
        betas = np.array(reg_model.params)
        x_list.append(ave_char)
        d_list.append(df_split)
        beta_list.append(betas)
    R = (beta_list[1] @ x_list[1]) - (beta_list[0] @ x_list[0])
    E = beta_list[1][1:] @ (x_list[1][1:] - x_list[0][1:])
    C = x_list[0][1:] @ (beta_list[1][1:] - beta_list[0][1:])
    U = beta_list[1][0] - beta_list[0][0]
    D = C + U
    idx = ['raw differential (R)', 'endowments (E)', 'coefficients (C)', 'constant (U)', 'discrimination (C+U)']
    data = np.transpose([[R,E,C,U, C+U],[R/R*100, E/R*100, C/R*100, U/R*100, (C+U)*100/R]])
    if abs(R - (E + C + U))<0.1:
        print('\nAverage Chinese ISEI: {:.5}'.format(d_list[1].isei.mean()))
        print('Average Malay ISEI    : {:.5}'.format(d_list[0].isei.mean()))
        print('\nDifference in average ISEI : {:.5}'.format(d_list[1].isei.mean() - d_list[0].isei.mean()))
        return pd.DataFrame(data, columns = ['level', '%'], index=idx)
    else:
        return [R, E+C+U]
```
